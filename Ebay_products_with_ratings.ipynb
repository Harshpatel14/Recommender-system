```python
df_pivot = data.pivot_table(index='Productid',columns='user_id',values='rating').fillna(0)
df_pivot
```

```python
from scipy.sparse import csr_matrix
df_pivot_matrix = csr_matrix(df_pivot.values)
print(df_pivot_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors

model_k = NearestNeighbors(metric = 'cosine',n_neighbors=25,radius=1)
model_k.fit(df_pivot_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_mat = cosine_similarity(df_pivot)
similarity_mat
```

```python
product_id = int(input("enter product id: "))
data1 = list(df_pivot.index)
print(data1)
```

```python
query_index = data1.index(product_id)
print(query_index)
```

```python
similarity, indices = model_k.kneighbors(df_pivot.iloc[query_index,:].values.reshape(1,-1),n_neighbors = 12)
print(similarity)
print(indices)
```

```python
data_dict={}
for i in range(0,len(similarity.flatten())):
    if i==0:
        print('Recommendation for {0}: \n'.format(df_pivot.index[query_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])] = float(similarity.flatten()[i])
        print(f'{df_pivot.index[indices.flatten()[i]]}, is similarity distance = with {similarity.flatten()[i]:.20f}: ')
print(data_dict)        
```
